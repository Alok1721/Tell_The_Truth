```python
import requests
from bs4 import BeautifulSoup
from transformers import pipeline

# Step 1: Fetch the content from the URL
def fetch_url_content(url):
    """
    Fetches and parses the main content of a given URL.
    :param url: The URL to scrape
    :return: The extracted text content or a message if scraping fails
    """
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise error for bad responses
        soup = BeautifulSoup(response.text, 'html.parser')
        paragraphs = soup.find_all('p')  # Extract text in <p> tags
        content = "\n".join(p.get_text() for p in paragraphs if p.get_text())
        return content.strip() if content else "No content found."
    except Exception as e:
        return f"An error occurred while fetching the URL: {e}"

# Step 2: Summarize the content
def summarize_content(content, summarizer):
    """
    Summarizes the given content using a pre-trained summarization pipeline.
    :param content: The text content to summarize
    :param summarizer: The pre-trained summarizer pipeline
    :return: The summarized text
    """
    try:
        # Limiting the content length for processing
        content = content[:1024]  # Many models have input limits
        summary = summarizer(content, max_length=100, min_length=30, do_sample=False)
        return summary[0]['summary_text'] if summary else "No summary generated."
    except Exception as e:
        return f"An error occurred during summarization: {e}"

# Step 3: Initialize the summarizer
def main():
    # List of URLs
    urls = [
        "https://en.wikipedia.org/wiki/Arvind_Kejriwal",
        "https://www.britannica.com/topic/Aam-Aadmi-Party",
        "https://en.wikipedia.org/wiki/Aam_Aadmi_Party",
        "https://www.britannica.com/biography/Arvind-Kejriwal",
        "https://www.indiatoday.in/india/story/arvind-kejriwal-resignation-delhi-cm-news-political-sixer-five-big-reasons-aam-aadmi-party-aap-delhi-assembly-election-2600535-2024-09-16",
        "https://www.jagranjosh.com/general-knowledge/delhi-chief-ministers-cm-list-party-name-and-tenure-1726561886-1",
        "https://www.bbc.com/news/articles/ce3wwjqpx20o",
        "https://indianexpress.com/article/cities/delhi/arvind-kejriwal-delhi-cm-tenure-highlights-9569057/",
        "https://in.linkedin.com/in/anandmangnale",
        "https://www.livemint.com/politics/arvind-kejriwals-d-j-vu-the-delhi-cms"
    ]

    # Load the summarization pipeline
    summarizer = pipeline("summarization")

    # Fetch and summarize content for each URL
    for idx, url in enumerate(urls, 1):
        print(f"\n{idx}. URL: {url}")
        print("-" * 80)

        content = fetch_url_content(url)
        if "error" in content.lower():
            print(content)
        else:
            summary = summarize_content(content, summarizer)
            print("Summary:")
            print(summary)

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import requests
from newspaper import Article
from sentence_transformers import SentenceTransformer, util
import numpy as np
import concurrent.futures

# Initialize the BERT model for sentence embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
news_article_text = "  The Internet Brutally Mocks Disney’s New Trump Robot At Hall Of Presidents"
query=news_article_text
print("\nquery is: ",query,"\n")

#it will fetch top 10 url from googleSearchEngine and store in fetched_url
fetched_url=[]
api_key = "AIzaSyA6x4oBkk1XiZcYFTr_PZTr5aHQP5gqbXU"
cx = "f21f1d598d24246ac"

def fetch_top_news(query, num_results=10):

    url = f"https://www.googleapis.com/customsearch/v1"
    params = {
        "q": query,
        "key": api_key,
        "cx": cx,

    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()

        results = []
        for item in data.get("items", []):
            title = item.get("title")
            link = item.get("link")
            results.append({"title": title, "link": link})
            fetched_url.append(link)

        return results

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")
        return []

news_results = fetch_top_news(query)
print("\n step3: lists of urls",fetched_url)

# Function to fetch content from a URL using Newspaper3k
def fetch_url_content_newspaper3k(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        return f"Error fetching content from URL: {e}"

# Function to compute similarity between news article text and content chunks
def compute_similarity(news_article_text, fetched_content_chunks):
    # Generate embedding for the news article
    news_embedding = model.encode(news_article_text, convert_to_tensor=True)

    # List to store similarity scores
    similarity_scores = []

    # Iterate over content chunks and compute similarity
    for chunk in fetched_content_chunks:
        chunk_embedding = model.encode(chunk, convert_to_tensor=True)
        similarity = util.pytorch_cos_sim(news_embedding, chunk_embedding).item()
        similarity_scores.append(similarity)

    # Return the highest similarity score
    return max(similarity_scores)

# Function to split content into chunks (e.g., sentences or paragraphs)
def split_content_into_chunks(content, chunk_type='sentence'):
    if chunk_type == 'sentence':
        # Split content into sentences
        sentences = nltk.sent_tokenize(content)
        return sentences
    elif chunk_type == 'paragraph':
        # Split content into paragraphs
        paragraphs = content.split('\n')
        return paragraphs
    else:
        return [content]  # Return whole content if no chunking method is provided

# Main method to process the content
def process_content(news_article_text, fetched_url_list):
    # Initialize list to store fetched content chunks
    all_content_chunks = []

    for idx, url in enumerate(fetched_url_list, 1):
        print(f"\n{idx}. URL: {url}")

        # Fetch the content from the URL
        content = fetch_url_content_newspaper3k(url)
        if "error" in content.lower():
            print(content)
        else:
            # Split content into chunks (sentences or paragraphs)
            content_chunks = split_content_into_chunks(content, chunk_type='sentence')  # Change to 'paragraph' if needed
            all_content_chunks.extend(content_chunks)

    # Compute the similarity for each chunk and return the highest similarity score
    similarity_score = compute_similarity(news_article_text, all_content_chunks)
    return similarity_score

# # Example usage:
# news_article_text = "Over 200 Flights Delayed, Train Ops Hit In Delhi As Dense Fog Reduces Visibility To Zero"
# fetched_url_list = [
#     'https://timesofindia.indiatimes.com/city/delhi/dense-fog-for-3-days-delhis-air-could-turn-severe/articleshow/106304823.cms',
#     'https://www.hindustantimes.com/cities/delhi-logs-13-hours-of-dense-fog-as-rail-flight-operations-take-a-hit-101703789642896.html',
#     'https://www.businesstoday.in/latest/trends/story/imd-predicts-very-dense-fog-over-north-india-till-jan-4-flight-train-operations-to-remain-affected-411378-2023-12-31'
# ]

# Process and compute similarity
# Function to process a URL and compute its similarity to the news article
def process_url(url):
    content = fetch_url_content_newspaper3k(url)
    if "error" in content.lower():
        return None
    else:
        # Split content into chunks (sentences or paragraphs)
        content_chunks = split_content_into_chunks(content, chunk_type='sentence')
        return content_chunks
# Main method to process the content concurrently
def process_content(news_article_text, fetched_url_list):
    all_content_chunks = []

    # Use parallel processing to fetch content concurrently from multiple URLs
    with concurrent.futures.ThreadPoolExecutor() as executor:
        future_to_url = {executor.submit(process_url, url): url for url in fetched_url_list}
        for future in concurrent.futures.as_completed(future_to_url):
            result = future.result()
            if result:
                all_content_chunks.extend(result)

    # Compute the similarity for each chunk and return the highest similarity score
    similarity_score = compute_similarity(news_article_text, all_content_chunks)
    return similarity_score

# Process and compute similarity
similarity_score = process_content(news_article_text, fetched_url)
print(f"\nHighest similarity score: {similarity_score}")

# Check if the news is real based on a threshold
if similarity_score > 0.4:
    print("\n\n\t\t-------------- News is Real --------------")
else:
    print("\n\n\t\t------------- News is Fake -------------")
```
